- å¹¶ä¸”æ»¡è¶³æœ€ç®€åŽŸåˆ™ï¼Œå³æ‰€é€‰è›‹ç™½æ•°é‡æœ€å°‘ï¼š

## 5.2 Protein-Level FDR Controlï¼ˆè›‹ç™½æ°´å¹³çš„ FDR æŽ§åˆ¶ï¼‰

åœ¨è‚½æ®µå±‚é¢æŽ§åˆ¶å‡é˜³æ€§ä¸è¶³ä»¥ä¿è¯è›‹ç™½å±‚é¢ç»“æžœçš„å¯é æ€§ã€‚

å› æ­¤ï¼Œéœ€è¦åœ¨è›‹ç™½è´¨æŽ¨æ–­åŽï¼Œå¯¹è›‹ç™½è¯†åˆ«ç»“æžœè¿›è¡Œ **å‡å‘çŽ°çŽ‡ï¼ˆFalse Discovery Rate, FDRï¼‰æŽ§åˆ¶**ã€‚

### å®žæ–½æ–¹æ³•ï¼š
- æž„å»ºè›‹ç™½è´¨çš„ Target-Decoy åˆ—è¡¨ï¼›
- é€šè¿‡**åˆ†æ•°æŽ’åº**ï¼ŒæŽ§åˆ¶è¯†åˆ«å‡ºçš„è›‹ç™½é›†åˆä¸­**å‡é˜³æ€§æ¯”ä¾‹ä¸è¶…è¿‡æŒ‡å®šé˜ˆå€¼**ï¼ˆå¦‚ 1%ï¼‰ã€‚

> âœ… é€šå¸¸ä¸Žè‚½æ®µçº§åˆ«çš„ FDR æŽ§åˆ¶è”åˆä½¿ç”¨ï¼Œæå‡æ•´ä½“ç»“æžœçš„å¯ä¿¡åº¦ã€‚

### è›‹ç™½æ°´å¹³çš„ FDR æŽ§åˆ¶ï¼šæ•°å­¦åŽŸç†

åœ¨ Target-Decoy æ–¹æ³•ä¸­ï¼š

- $\mathcal{T}$ï¼šè¢«è¯†åˆ«çš„ target è›‹ç™½æ•°é‡ï¼ˆçœŸå®žæ•°æ®åº“ä¸­çš„è›‹ç™½ï¼‰ï¼›
- $\mathcal{D}$ï¼šè¢«è¯†åˆ«çš„ decoy è›‹ç™½æ•°é‡ï¼ˆè™šæž„æ•°æ®åº“ä¸­çš„è›‹ç™½ï¼‰ï¼›
- æ‰€æœ‰è›‹ç™½æŒ‰æ‰“åˆ†ï¼ˆscoreï¼‰ä»Žé«˜åˆ°ä½ŽæŽ’åºï¼Œè®¾å®šä¸€ä¸ªæ‰“åˆ†é˜ˆå€¼ $s$ï¼›
- ç»Ÿè®¡åœ¨è¯¥é˜ˆå€¼ $s$ ä»¥ä¸Šçš„ target å’Œ decoy æ•°é‡ã€‚

å®šä¹‰ FDRï¼ˆFalse Discovery Rateï¼‰ä¸ºï¼š

$$
\text{FDR}(s) = \frac{|\mathcal{D}_s|}{|\mathcal{T}_s|}
$$

å…¶ä¸­ï¼š
- $\mathcal{D}_s = \{d \in \text{decoy proteins} \mid \text{score}(d) \geq s \}$ï¼›
- $\mathcal{T}_s = \{t \in \text{target proteins} \mid \text{score}(t) \geq s \}$ã€‚


#### ç›®æ ‡ï¼š
æ‰¾åˆ°æ»¡è¶³ FDR ä¸è¶…è¿‡é˜ˆå€¼ $\alpha$ï¼ˆé€šå¸¸ä¸º 0.01ï¼‰çš„æœ€å¤§æ‰“åˆ†é˜ˆå€¼ $s^*$ï¼š

$$
s^* = \max \left\{ s \,\middle|\, \frac{|\mathcal{D}_s|}{|\mathcal{T}_s|} \leq \alpha \right\}
$$

ç„¶åŽä¿ç•™æ‰€æœ‰ $\text{score} \geq s^*$ çš„è›‹ç™½ä½œä¸ºæœ€ç»ˆå¯ä¿¡è¯†åˆ«ç»“æžœã€‚



```python
!pip install pandas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# æ¨¡æ‹Ÿæ•°æ®ï¼š1000ä¸ª target + 1000ä¸ª decoy
np.random.seed(42)
target_scores = np.random.normal(loc=50, scale=10, size=1000)
decoy_scores = np.random.normal(loc=40, scale=10, size=1000)

# æž„å»º DataFrame
df = pd.DataFrame({
    'score': np.concatenate([target_scores, decoy_scores]),
    'type': ['target'] * 1000 + ['decoy'] * 1000
})

# æŒ‰æ‰“åˆ†ä»Žé«˜åˆ°ä½ŽæŽ’åº
df = df.sort_values(by='score', ascending=False).reset_index(drop=True)

# ç´¯è®¡ target / decoy æ•°é‡
df['cum_target'] = (df['type'] == 'target').cumsum()
df['cum_decoy'] = (df['type'] == 'decoy').cumsum()

# FDR è®¡ç®—
df['FDR'] = df['cum_decoy'] / df['cum_target']

# è®¾å®šé˜ˆå€¼ï¼Œä¾‹å¦‚ FDR <= 1%
threshold_index = df[df['FDR'] <= 0.01].first_valid_index()
score_threshold = df.loc[threshold_index, 'score']
print(f"Score threshold at 1% FDR: {score_threshold:.2f}")

# å¯è§†åŒ–
plt.figure(figsize=(8, 5))
plt.plot(df['score'], df['FDR'], label='FDR curve')
plt.axhline(0.01, color='red', linestyle='--', label='1% FDR')
plt.axvline(score_threshold, color='green', linestyle='--', label=f'Threshold = {score_threshold:.2f}')
plt.xlabel('Score')
plt.ylabel('FDR')
plt.title('Protein-level FDR vs. Score Threshold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 5.3 è›‹ç™½ç»„å®šé‡çš„æ•°å­¦å»ºæ¨¡

è®¾ï¼š
- ä¸€ä¸ªè›‹ç™½ç»„åŒ…å«è›‹ç™½ $P_1, P_2, \dots, P_K$ï¼›
- è¢«è¯†åˆ«å‡ºçš„è‚½æ®µé›†åˆä¸º $\mathcal{P} = \{p_1, p_2, \dots, p_n\}$ï¼›
- æ¯ä¸ªè‚½æ®µ $p_i$ çš„å®šé‡å¼ºåº¦ä¸º $I_i$ï¼Œä¸” $p_i$ å¯ç”±è›‹ç™½é›†åˆ $S_i \subseteq \{P_1, \dots, P_K\}$ ç”Ÿæˆã€‚

#### 1. å”¯ä¸€è‚½æ®µæ³•ï¼ˆUnique Peptides Onlyï¼‰

- æå– $\mathcal{U} = \{p_i \mid |S_i| = 1 \}$ï¼ˆåªæ¥è‡ªä¸€ä¸ªè›‹ç™½ç»„çš„è‚½æ®µï¼‰ï¼›
- å®šé‡å€¼ä¸ºå”¯ä¸€è‚½æ®µçš„å¹³å‡æˆ–åŠ æƒå’Œï¼š

$$
I_{\text{protein group}} = \frac{1}{|\mathcal{U}|} \sum_{p_i \in \mathcal{U}} I_i
\quad \text{æˆ–} \quad
I_{\text{weighted}} = \sum_{p_i \in \mathcal{U}} w_i I_i
$$

å…¶ä¸­ $w_i$ æ˜¯ç½®ä¿¡åº¦æˆ–é‡å¤æ€§ç›¸å…³çš„æƒé‡ã€‚


#### 2. åŠ æƒå¹³å‡æ³•ï¼ˆWeighted Mean of All Peptidesï¼‰

- ä¸è®ºè‚½æ®µæ˜¯å¦å”¯ä¸€ï¼Œå…¨éƒ¨çº³å…¥ï¼ŒæŒ‰è‚½æ®µè´¨é‡ã€ç¨³å®šæ€§ç­‰èµ‹æƒï¼š

\quad \text{with} \quad \sum w_i = 1

- ç¤ºä¾‹ï¼šå¯æ ¹æ®è‚½æ®µåœ¨æ‰€æœ‰æ ·æœ¬ä¸­çš„å˜å¼‚ç³»æ•°ï¼ˆCVï¼‰æ¥è°ƒæ•´ $w_i$ã€‚


#### 3. å…±äº«è‚½æ®µåˆ†é…æ³•ï¼ˆShared Peptide Modelingï¼‰

- è®¾æ¯ä¸ªè›‹ç™½ $P_j$ çš„æœªçŸ¥è¡¨è¾¾å¼ºåº¦ä¸º $x_j$ï¼›
- è‚½æ®µå¼ºåº¦ä¸ºè¿™äº›è›‹ç™½çš„çº¿æ€§ç»„åˆï¼š

$$
I_i = \sum_{j: P_j \in S_i} x_j
$$

- æ‰€æœ‰è‚½æ®µå½¢æˆä¸€ä¸ªçº¿æ€§æ–¹ç¨‹ç»„ï¼š

$$

\begin{bmatrix}

=
\begin{bmatrix}

\end{bmatrix}
$$

- ç”¨æœ€å°äºŒä¹˜æ³•æˆ–æ­£åˆ™åŒ–ï¼ˆå¦‚ Ridge/Lassoï¼‰ä¼°è®¡æ¯ä¸ªè›‹ç™½çš„è¡¨è¾¾é‡ã€‚


#### å…±äº«è‚½æ®µåˆ†é…æ³•ä¸­çš„çŸ©é˜µæž„é€ è§£é‡Š

æˆ‘ä»¬å¸Œæœ›æ ¹æ®è‚½æ®µä¸Žè›‹ç™½ä¹‹é—´çš„æ˜ å°„å…³ç³»ï¼Œå»ºç«‹çº¿æ€§æ¨¡åž‹ï¼š

##### è®¾å®šï¼š

- è›‹ç™½è´¨ï¼š$P_1, P_2, P_3$
- è‚½æ®µï¼š$p_1, p_2, p_3$
- æ¯ä¸ªè‚½æ®µçš„å¼ºåº¦ä¸º $I_1, I_2, I_3$
- æ¯ä¸ªè›‹ç™½çš„æœªçŸ¥è¡¨è¾¾é‡ä¸º $x_1, x_2, x_3$

##### å‡è®¾æ¥æºå…³ç³»å¦‚ä¸‹ï¼š

- $p_1$ æ¥æºäºŽ $P_1$ å’Œ $P_2$
- $p_2$ æ¥æºäºŽ $P_2$ å’Œ $P_3$
- $p_3$ æ¥æºäºŽ $P_1$

##### åˆ™è®¾è®¡çŸ©é˜µä¸ºï¼š

$$
A =
\begin{bmatrix}
1 & 1 & 0 \\

$$

##### ç¤ºä¾‹éªŒè¯ï¼š

è‹¥ $x_1 = 10$, $x_2 = 5$, $x_3 = 8$ï¼Œåˆ™ï¼š

$$
\begin{aligned}

```python
import numpy as np
from numpy.linalg import lstsq

# æž„å»ºè®¾è®¡çŸ©é˜µ Aï¼ˆè‚½æ®µ Ã— è›‹ç™½ï¼‰
A = np.array([
    [1, 1, 0],  # p1 from P1 and P2
    [0, 1, 1],  # p2 from P2 and P3
    [1, 0, 0]   # p3 from P1
])

# è‚½æ®µè§‚æµ‹å¼ºåº¦å‘é‡ï¼ˆå‡è®¾å®žéªŒæµ‹åˆ°ï¼‰
I = np.array([15, 13, 10])  # I1, I2, I3

# æ±‚è§£æœªçŸ¥è›‹ç™½è¡¨è¾¾é‡ x1, x2, x3
x, residuals, rank, s = lstsq(A, I, rcond=None)

# è¾“å‡ºç»“æžœ
print(f"Estimated protein intensities:\n x1 = {x[0]:.2f}, x2 = {x[1]:.2f}, x3 = {x[2]:.2f}")
```

> åœ¨æ‰€æœ‰è¢«è®¤ä¸ºâ€œæ˜¾è‘—â€çš„æ£€éªŒç»“æžœä¸­ï¼Œæœ€å¤šæœ‰ $\alpha$ æ¯”ä¾‹æ˜¯é”™è¯¯å‘çŽ°ï¼ˆå‡é˜³æ€§ï¼‰ã€‚

### 2. åŠŸèƒ½å¯Œé›†åˆ†æžï¼ˆEnrichment Analysisï¼‰

#### ðŸ“‰ ä¸ºä»€ä¹ˆè¦çœ‹â€œè‡³å°‘è¿™ä¹ˆæžç«¯â€ï¼ˆRight-tailï¼‰

```python
from scipy.stats import hypergeom

# è®¾ç½®å‚æ•°ï¼š
M = 20000   # èƒŒæ™¯æ€»è›‹ç™½æ•°ï¼ˆuniverse sizeï¼‰
K = 200     # æŸé€šè·¯ä¸­è›‹ç™½æ•°ï¼ˆè¯¥é€šè·¯å« K ä¸ªè›‹ç™½ï¼‰
n = 300     # ä½ çš„å·®å¼‚è›‹ç™½æ•°é‡
k = 15      # å·®å¼‚è›‹ç™½ä¸­å‡ºçŽ°åœ¨è¯¥é€šè·¯ä¸­çš„æ•°é‡

# è®¡ç®—å³å°¾ p å€¼ï¼šè‡³å°‘è§‚å¯Ÿåˆ° k ä¸ªé€šè·¯è›‹ç™½çš„æ¦‚çŽ‡
# æ³¨æ„ï¼šsf = survival function = P(X >= k)
p_value = hypergeom.sf(k - 1, M, K, n)

print(f"P(X â‰¥ {k}) = {p_value:.4e}")
if p_value < 0.05:
    print("âœ… æ˜¾è‘—å¯Œé›†ï¼šé€šè·¯å¯èƒ½åœ¨å®žéªŒæ¡ä»¶ä¸‹å‘ç”Ÿè°ƒæŽ§")
else:
    print("âŒ ä¸æ˜¾è‘—ï¼šæ— æ³•æŽ’é™¤è¿™æ˜¯éšæœºçŽ°è±¡")
```
